1. Genera una función llamada leer_tabla(tabla, engine) y utilízala para leer tablas

```python
df_OrderDetails = fn.leer_tabla('orderdetails', engine)
#df_OrderDetails
df_Orders = fn.leer_tabla('orders', engine)
# df_Orders 
df_Customers = fn.leer_tabla('customers', engine)
#df_Customers
df_Products = fn.leer_tabla('products', engine)
#df_Products
df_Employees = fn.leer_tabla('employees', engine)
#df_Employees
```

2. Realiza el cruce entre los DataFrames, asegurándote de utilizar correctamente el
parámetro validate para asegurar la integridad referencial
 tablas
 - order
 - orderdetails
 - customers
 - products
 - employees

```python
# union de tabla detalle de ordenes y encabezado de ordenes
df_Base=df_OrderDetails.merge(df_Orders, on='orderNumber', how='left', validate = 'many_to_one')
# ahora uno a la base anterior el maestro de productos
df_Base=df_Base.merge(df_Products, on='productCode', how='left', validate = 'many_to_one')
# ahora uno a la base anterior el maestro de compradores
df_Base=df_Base.merge(df_Customers, on='customerNumber', how='left', validate = 'many_to_one')
# ahora uno a la base anterior el maestro de empleados
df_Base=df_Base.merge(df_Employees, left_on='salesRepEmployeeNumber', right_on='employeeNumber', how='left', validate = 'many_to_one')
#revision
#print(df_Base.head())
#print(df_Base.dtypes)
```

3. Agrega las siguientes columnas, considerando su nombre y la fórmula asociada
- venta: quantityOrdered*priceEach
- costo: quantityOrdered*buyPrice
- ganancia: considerando las columnas anteriores


```python
df_Base['Venta'] = df_Base['quantityOrdered'] * df_Base['priceEach']
df_Base['Costo'] = df_Base['quantityOrdered'] * df_Base['buyPrice']
df_Base['Ganancia'] = df_Base['Venta'] - df_Base['Costo']
#revision
#print(df_Base.head())
#print(df_Base.dtypes)
```
